```python
import pandas as pd
from sklearn.model_selection import train_test_split


file_path = r"C:\Users\Abhishek\Downloads\Dataset - Form responses 1 (4).csv"
df = pd.read_csv(file_path)


print("Original Data:")
print(df.head())


df['ROLL NUMBER'] = pd.to_numeric(df['ROLL NUMBER'], errors='coerce')
df['MARKS'] = pd.to_numeric(df['MARKS'], errors='coerce')
df['STIPENED'] = pd.to_numeric(df['STIPENED'], errors='coerce')


df['MARKS'].fillna(df['MARKS'].mean(), inplace=True)


df['ROLL NUMBER'].fillna(0, inplace=True)


df['GENDER'].fillna('UNKNOWN', inplace=True)


print("Cleaned Data:")
print(df.head())


train, test = train_test_split(df, test_size=0.2, random_state=42)


print("Training Data:")
print(train.head())
print("Testing Data:")
print(test.head())
```
