```python
import pandas as pd

# Use raw string literal to avoid unicode escape error
file_path = r'C:\Users\Abhishek\Desktop\Alpha Intern\WorkSheet1.5\Dataset - Form responses 1 (3).csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Display the column names to find the correct names
print(df.columns)

# Assuming the correct names are 'MARKS' and 'STIPENED'
# Convert empty strings to NaN
df.replace('', float('nan'), inplace=True)

# Convert 'MARKS' and 'STIPENED' columns to numeric
df['MARKS'] = pd.to_numeric(df['MARKS'], errors='coerce')
df['STIPENED'] = pd.to_numeric(df['STIPENED'], errors='coerce')

# Calculate the interquartile range (IQR)
Q1 = df[['MARKS', 'STIPENED']].quantile(0.25)
Q3 = df[['MARKS', 'STIPENED']].quantile(0.75)
IQR = Q3 - Q1

# Identify outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = ((df[['MARKS', 'STIPENED']] < lower_bound) | (df[['MARKS', 'STIPENED']] > upper_bound)).any(axis=1)

# Remove outliers
cleaned_df = df[~outliers]

# Display the cleaned DataFrame
cleaned_df
```
